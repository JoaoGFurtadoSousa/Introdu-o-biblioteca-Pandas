```python
# Testando a pessoa que tem mais pontos
# Aqui ele esta pegando o maior valor da nossa series e armazenando na nossa variavel maior
# Após isso, estamos colocando na variavel condição todos os dados que sao falsos.
# na ultima linha de codigo ele vai exibir a unica linha que deu TRUE e mostrando o nome dele
maiorP=dataframe['Points'].max()
condicao=dataframe['Points']==maiorP
dataframe[condicao]['Name'].iloc[0]
```

```python
# Fazendo com a menor

menor=dataframe['Points'].min()
condicaom=dataframe['Points']==menor
dataframe[condicaom]
```

```python
# Vendo quem esta entre mil e dois mil pontos

meio=(dataframe['Points']>1000) & (dataframe['Points']<2000)
dataframe[meio]
```
